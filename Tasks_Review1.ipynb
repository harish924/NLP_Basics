# Frequency distribution and Conditional Frequency distribution

```python
#Conditional frequency distribution is collection of frequency distributions which means results produced based on some condition applied
from nltk.probability import ConditionalFreqDist
speech = inaugural.words(fileids="1937-Roosevelt.txt")
cfd = ConditionalFreqDist((len(word), word) for word in speech)
cfd[3]
```

```python
#four letter words
cfd[4]
```

```python
#five letter words
cfd[5]
```

```python
cfd[6]
```

```python
#frequency distribution of all the  words which are greater than 4 letters in the corpus
freq_words = []
for words in speech:
    if len(words)>4:
        freq_words.append(words)
```

```python
fd = nltk.FreqDist(freq_words)
```

```python
fd
```

```python
#here we are sorting according to frequency so the last one will be the highest times repeated(asscending order)
sort_fd= sorted(fd.items(), key = lambda x:x[1])
```

```python
fw = list(sort_fd)[len(sort_fd)-1]
```

```python
print("Most Used word is :",fw)
```

# Stemming paragraphs

```python
#porterstemmer which is the most used stemmer and it is least aggresive
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
example = "The reason doctors don’t give patients specific prognoses is not merely because they cannot.What patients seek is not scientific knowledge that doctors hide but existential authenticity each person must find on her own. Getting too deeply into statistics is like trying to quench a thirst with salty water. The angst of facing mortality has no remedy in probability."
example = [stemmer.stem(token) for token in example.split(" ")]
print(" ".join(example))
```

```python
#lancaster stemmer is more sensitive and becomes difficult to understand small words when applied on them
stemmer = LancasterStemmer()
example1 = "The reason doctors don’t give patients specific prognoses is not merely because they cannot.What patients seek is not scientific knowledge that doctors hide but existential authenticity each person must find on her own. Getting too deeply into statistics is like trying to quench a thirst with salty water. The angst of facing mortality has no remedy in probability."
example1 = [stemmer.stem(token) for token in example1.split(" ")]
print(" ".join(example1))
```

```python
#lemmatizer can be used to understand the meaning of a word if additional information such as pos is provided
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("cacti"))
print(lemmatizer.lemmatize("mice"))
print(lemmatizer.lemmatize("rocks"))
print(lemmatizer.lemmatize("better",pos='a'))
print(lemmatizer.lemmatize("Am")) 
print(lemmatizer.lemmatize("is", pos = 'v'))
```

# Basic Text Processing

```python
#tokenization of words from the sentences
import nltk
statement = "This Corona virus is going to kill us all"
words = nltk.word_tokenize(statement)
print(words)
```

```python
texts = ["""German prosecutors are investigating whether the prime suspect in the disappearance of Madeleine McCann was involved in a similar case of a five-year-old girl who went missing in Germany in 2015."""]
for text in texts:
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
    print(words)
```

```python
tagged = nltk.pos_tag(words)
print(tagged)
```

# Parts of Speech Tagging
